```python
### write a function that takes in the actual and budget dataframes and returns the percentage change in net profit, gross sales, net revenue and volume
def percentage_change_budget(actual, budget):
    actual = actual.groupby('Product').agg({'Net profit': 'sum', 'Gross Sales': 'sum', 'Net revenue': 'sum', 'Volume': 'sum'})
    budget = budget.groupby('Product').agg({'Net profit': 'sum', 'Gross Sales': 'sum', 'Net revenue': 'sum', 'Volume': 'sum'})
    actual['Net profit'] = actual['Net profit'].astype(float)
    actual['Gross Sales'] = actual['Gross Sales'].astype(float)
    actual['Net revenue'] = actual['Net revenue'].astype(float)
    actual['Volume'] = actual['Volume'].astype(float)
    budget['Net profit'] = budget['Net profit'].astype(float)
    budget['Gross Sales'] = budget['Gross Sales'].astype(float)
    budget['Net revenue'] = budget['Net revenue'].astype(float)
    #budget['Volume'] = budget['Volume'].astype(float)
    percentage_change = pd.DataFrame()
    percentage_change['Net profit'] = ((actual['Net profit'] - budget['Net profit']) / budget['Net profit']) * 100
    percentage_change['Gross Sales'] = ((actual['Gross Sales'] - budget['Gross Sales']) / budget['Gross Sales']) * 100
    percentage_change['Net revenue'] = ((actual['Net revenue'] - budget['Net revenue']) / budget['Net revenue']) * 100
    #percentage_change['Volume'] = ((actual['Volume'] - budget['Volume']) / budget['Volume']) * 100
    return percentage_change
```

```python
### write a function that takes in the actual and budget dataframes and returns the percentage change in net profit, gross sales, net revenue and volume
def percentage_change(actual, budget):
    actual = actual.groupby('Product').agg({'Net profit': 'sum', 'Gross Sales': 'sum', 'Net revenue': 'sum', 'Volume': 'sum'})
    budget = budget.groupby('Product').agg({'Net profit': 'sum', 'Gross Sales': 'sum', 'Net revenue': 'sum', 'Volume': 'sum'})
    actual['Net profit'] = actual['Net profit'].astype(float)
    actual['Gross Sales'] = actual['Gross Sales'].astype(float)
    actual['Net revenue'] = actual['Net revenue'].astype(float)
    actual['Volume'] = actual['Volume'].astype(float)
    budget['Net profit'] = budget['Net profit'].astype(float)
    budget['Gross Sales'] = budget['Gross Sales'].astype(float)
    budget['Net revenue'] = budget['Net revenue'].astype(float)
    budget['Volume'] = budget['Volume'].astype(float)
    percentage_change = pd.DataFrame()
    percentage_change['Net profit'] = ((actual['Net profit'] - budget['Net profit']) / budget['Net profit']) * 100
    percentage_change['Gross Sales'] = ((actual['Gross Sales'] - budget['Gross Sales']) / budget['Gross Sales']) * 100
    percentage_change['Net revenue'] = ((actual['Net revenue'] - budget['Net revenue']) / budget['Net revenue']) * 100
    percentage_change['Volume'] = ((actual['Volume'] - budget['Volume']) / budget['Volume']) * 100
    return percentage_change
```

## Plotting

```python
import altair as alt
```

```python
plot_budget = sep_2023_budget.reset_index().drop(columns='Volume')
plot_actual = sep_2023_actual.reset_index().drop(columns='Volume')
df_combined = pd.concat([plot_budget.set_index('Product').stack(), plot_actual.set_index('Product').stack()], axis=1, keys=['Budget', 'Actual'])


df_combined.index.names = ['Product', 'level_1']
df_combined = df_combined.reset_index()
df_combined
```

```python
df_melted = pd.melt(df_combined, id_vars=['Product', 'level_1'], 
                    var_name='Metric', value_name='Value')
df_melted = df_melted.rename(columns={'Metric': 'level_1', 'level_1': 'Metric'}) 
df_melted
```

```python
chart_2023_budget_actual = alt.Chart(df_melted).mark_bar().encode(
    x=alt.X('Product:N', title=None),
    xOffset='level_1:N',
    y=alt.Y('Value:Q', title=None),
    color=alt.Color('level_1:N', legend=alt.Legend(title='Legend')),  
    column=alt.Column('Metric:N', title=None),
    order=alt.Order('level_1:N', sort='descending')  
)

chart_2023_budget_actual.properties(title=alt.Title('Sep 2023 Budget vs Actual by Product', anchor='middle'), width=300)
```
